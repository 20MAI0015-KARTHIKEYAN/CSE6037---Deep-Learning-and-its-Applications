```python
#importing the required modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.linspace(-6, 6)
```

# Sigmoid
f(z) = 1/(1+exp(-z))

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
plt.plot(z,sigmoid(z))
plt.title('Sigmoid Activation Function')
plt.show()
```

# tanh
fn(z) = (1-exp(-2z))/(1+exp(-2z))

```python
def tanh(z):
    return np.tanh(z)
plt.plot(z, tanh(z))
plt.title('tanh Activation Function')
plt.show()
```

# Relu
f(z) = max(z,0)

```python
def relu(z):
    z1=[]
    for i in z:
        if i<0:
            z1.append(0)
        else:
            z1.append(i)
    return z1

plt.plot(z, relu(z))
plt.title('Activation Function :RELU')
plt.show()
```

# Leaky-Relu
fn(z) = max(0.01*z,z)

```python
def lrelu(z):
  if z > 0:
    return z
  else:
    return z*0.01

plt.plot(z, list(map(lambda z: lrelu(z),z)))
plt.title('Leaky-Relu Activation Function')
plt.show()
```

# Exponential Learning Unit
f(z) = z if z > 0, else alpha((exp(z))−1)

```python
def elu_prime(z,alpha):
    return 1 if z > 0 else alpha*np.exp(z)
```

```python
plt.plot(z, list(map(lambda z: elu_prime(z,0.1),z)))
plt.title('ELU Activation Function')
plt.show()
```

```python
def elu(z):
    if z > 0 :
        return z
    else : 
        return (np.exp(z)-1)
```

```python
plt.plot(z, list(map(lambda z: elu(z),z)))
plt.title('ELU Activation Function')
plt.show()
```

```python
# importing the requirements
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Conv2D
import skimage.measure
```

```python
from PIL import Image
a = Image.open(r'shop.jpg')
```

```python
plt.imshow(a)
```

```python
r_size = a.resize((np.array((100,100))).astype(int))
```

```python
b = np.asarray(r_size)
```

```python
b.shape
x=b.reshape(1,100,100,3) 
```

```python
def Convolution(x,kernel_size,strides=1,padding='valid'):
    model=Sequential()
    model.add(Conv2D(1,kernel_size=kernel_size,strides=strides,padding=padding,input_shape=(100,100,3)))
    yhat=model.predict(x)
    y=yhat.reshape(yhat.shape[1],yhat.shape[2])
    filtered_image=Image.fromarray(y)
    return filtered_image
```

1. filter size (3*3)

```python
img1=Convolution(x,(3,3))
plt.imshow(img1)
entropy = skimage.measure.shannon_entropy(img1)
entropy
```

2. filter size (5*5)

```python
img2=Convolution(x,(5,5))
plt.imshow(img2)
entropy = skimage.measure.shannon_entropy(img2)
entropy
```

3. filter size (3*3), stride=2

```python
img3=Convolution(x,(3,3),2)
plt.imshow(img3)
entropy = skimage.measure.shannon_entropy(img3)
entropy
```

4. filter size (5*5), stride=2

```python
img4=Convolution(x,(3,3),2)
plt.imshow(img4)
entropy = skimage.measure.shannon_entropy(img4)
entropy
```

5. filter size (3*3),stride=1, zero-padding

```python
img5=Convolution(x,(3,3),1,'same')
plt.imshow(img5)
entropy = skimage.measure.shannon_entropy(img5)
entropy
```

6. filter size (5*5),stride=1, zero-padding
